## What is Language AI?

* **Formal Definition**: AI pioneer John McCarthy defined AI as the science and engineering of creating intelligent machines, especially computer programs, which may or may not follow biologically inspired methods.

* **Overuse of the Term**: The term "AI" is often misapplied to systems that don’t truly demonstrate intelligence, such as basic if-else logic in video game characters (NPCs).

* While **Large Language Models (LLMs)** are a major focus in Language AI, the field also includes other impactful technologies like **retrieval systems** that enhance LLMs.

## Representation Models (Encoder Only Models)

1. **BERT Introduction (2018)**:
   BERT (Bidirectional Encoder Representations from Transformers) was introduced as a **transformer-based, encoder-only architecture**, unlike the original Transformer which used an encoder-decoder setup for translation tasks.

2. **Encoder-Only Architecture**:
   BERT removes the decoder component and **focuses only on encoding input text**, making it ideal for understanding and analyzing language rather than generating it.

3. **Use of \[CLS] Token**:
   A special token `[CLS]` is added at the start of input sequences. Its final embedding is used for tasks like **classification**, summarizing the meaning of the entire input.

4. **Masked Language Modeling (MLM)**:
   BERT is trained using MLM—**randomly masking words in a sentence** and training the model to predict them. This helps it learn **contextual language understanding**.

5. **Pretraining and Fine-tuning**:
   BERT is first **pretrained on large corpora** (e.g., Wikipedia) and then **fine-tuned** on specific tasks (like sentiment analysis or question answering), requiring **less data and compute**.

6. **Feature Extraction**:
   BERT models can also be used without fine-tuning to **extract embeddings** (vector representations) of text, making them useful for downstream tasks like clustering or semantic search.

7. **Wide Range of Applications**:
   BERT and its variants are used in tasks such as **text classification**, **semantic search**, and **clustering**, and remain a core tool in NLP.

8. **Representation Models vs. Generative Models**:

   * **Representation models** (like BERT): Focus on understanding and creating vector representations of text.
   * **Generative models** (like GPT): Focus on producing text outputs (e.g., answering questions or writing essays).

9. **Long-term Impact**:
    BERT laid the foundation for **modern NLP models** and significantly advanced the use of **transfer learning in language AI**.


## Generative Models (Decoder only Models)

1. **Decoder-Only Architecture**:
   Introduced in 2018 with **GPT-1**, decoder-only models focus on **generating** language rather than understanding it. They **stack decoder blocks** and remove encoder components.

2. **Pretraining Data**:
   GPT-1 was trained on 7,000 books and the **Common Crawl** dataset, resulting in a model with **117 million parameters**.

3. **Rapid Scaling of Model Size**:
   The number of parameters grew significantly in later versions:

   * **GPT-2**: 1.5 billion parameters
   * **GPT-3**: 175 billion parameters
     This scaling led to **major improvements** in language generation.

4. **LLMs and Decoder-Only Models**:
   These large models are commonly called **Large Language Models (LLMs)**. Though often associated with generation, the term also includes **encoder-only models** like BERT.

5. **Generative Capabilities**:
   These models **autocomplete text** given an input. This core function allows them to be used for tasks like text generation, summarization, and answering questions.

6. **From Completion to Chatbots**:
   Fine-tuning decoder-only models enables them to function as **chatbots or instruction-following models**, capable of understanding user queries and generating relevant responses.

7. **Prompt-Based Completion**:
   These models are often described as **completion models**, meaning they generate text that logically follows the input **prompt**.

8. **Context Length / Window**:
   The **context window** defines how many tokens the model can handle at once. A **larger window** enables the model to process long documents or conversations more effectively.

9. **Autoregressive Nature**:
   Decoder-only models generate text **token by token**, where each new token increases the current context, building the output sequentially.


## Year of GenAI

**1. 2023: The Year of Generative AI**
The release and widespread adoption of ChatGPT (based on GPT-3.5) triggered massive public and industry interest in generative AI. This led many to label 2023 as the “Year of Generative AI.”

**2. ChatGPT: Product vs Model**
Although often used interchangeably, *ChatGPT* is the product interface, not the actual model. Initially powered by GPT-3.5, it later integrated more powerful models like GPT-4, reflecting rapid evolution in model performance.

**3. Rise of Proprietary Models**

**Proprietary LLMs**—like GPT-3.5/GPT-4 (OpenAI), Claude (Anthropic), and Gemini (Google)—are:

* **Closed-source and commercial**: Users access them via APIs or platforms.
* **Highly optimized**: Trained using large-scale resources and fine-tuned for safety, instruction-following, and broad use cases.
* **Business-friendly**: They offer security, stability, and customer support but are restricted in customization and transparency.

**4. Growth of Open-Source Models**

**Open-source models**, like LLaMA, Mistral, and Falcon, saw major adoption due to:

* **Transparency**: Code and model weights are public, promoting trust and research.
* **Customizability**: Developers can fine-tune them for specific domains or languages.
* **Community-driven innovation**: Open collaboration enables rapid improvements, though users must manage hosting and safety concerns themselves.

**5. Emergence of New Architectures**
Beyond Transformers (which power most LLMs), new architectures like **Mamba** and **RWKV** emerged. These aim to:

* **Improve efficiency**: With faster inference and larger context handling.
* **Expand possibilities**: Making AI accessible on edge devices or lower-cost infrastructure.

**6. Explosion of Model Development**
Both open-source and proprietary models were released at an unprecedented pace, making it difficult to keep up. This reflects the fast-moving nature of the field and the intense innovation competition among AI labs.

**7. Broader Focus Beyond LLMs**
The AI ecosystem in 2023 wasn’t limited to LLMs. Techniques like **embedding models**, **encoder-only architectures** (e.g., BERT), and even traditional methods like **bag-of-words** remained valuable, especially in hybrid systems supporting or enhancing LLM functionality.


## Moving Definition of LLM

1. **The term “Large Language Model” (LLM) is evolving and inconsistent** — Traditionally, it referred to large, generative, decoder-only Transformer models (like GPT-3/4), but this definition is too narrow. Smaller models with similar capabilities or large models used for non-generative tasks (like classification) challenge this label.

2. **“Large” is relative, and usefulness matters more than size** — The classification should focus on **capabilities**, not just parameter count or generative ability. Even smaller or non-generative models (like embedding or representation models) can meaningfully contribute to LLM applications and will be covered in the book.

## The Training Paradigm of LLM


1. **Traditional ML vs LLMs**: Traditional machine learning involves a single-step process where a model is trained directly for a specific task (e.g., classification or regression).

2. **Two-Step Training for LLMs**: LLM training generally involves two main phases—**pretraining** and **fine-tuning**—making it more complex and resource-intensive than traditional ML.

3. **Pretraining (Language Modeling)**: In this first phase, the model is trained on massive internet-scale text data to learn general language patterns like grammar, context, and word prediction. This produces a **base or foundation model** that isn't yet task-specific.

4. **Fine-Tuning (Task Adaptation)**: The pretrained model is then refined on smaller, task-specific datasets to make it follow instructions or perform specialized tasks like classification, summarization, or question answering.

5. **Efficiency of Fine-Tuning**: Fine-tuning is much cheaper than pretraining and makes powerful models accessible to more users and organizations, despite the enormous resources required for the initial pretraining.

6. **Example – Llama 2**: Llama 2 was pretrained on **2 trillion tokens**, illustrating the massive scale of modern LLM training and the need for pretraining to be done by well-resourced organizations.

7. **Ongoing Adaptation**: Additional fine-tuning or post-training methods (discussed later in the book) can further align the model to user needs, preferences, or specific behaviors.


## Responsible LLM Development and Usage

1. **Bias and Fairness**:
   LLMs can absorb and **amplify biases** from their training data, which is often undisclosed—making it hard to detect or correct unfair behaviors without real-world testing.

2. **Transparency and Accountability**:
   LLMs may be **indistinguishable from humans** in conversations, leading to risks in sensitive domains (e.g., healthcare), where **human oversight** is critical.

3. **Harmful Content Generation**:
   LLMs can produce **false or misleading information** confidently, and may be misused to generate **fake news or deceptive content**.

4. **Intellectual Property Issues**:
   There's legal uncertainty around who owns LLM outputs—**the user, the model creator, or the original author** of training data, especially if content overlaps with copyrighted material.

5. **Need for Regulation**:
   Governments are starting to **regulate LLMs** due to their growing influence. For instance, the **European AI Act** governs foundation models and their deployment.

6. **Ethical Responsibility**:
   Developers and users are encouraged to prioritize **ethical, safe, and responsible AI use**, keeping in mind the societal impact of these systems.

7. **Lack of Ground Truth**: LLMs can **hallucinate** facts—they do not    guarantee truthfulness, even when they sound confident and fluent.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load the Phi-3 Mini model for causal language modeling with GPU support and auto dtype
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",              # Automatically place model on GPU
    torch_dtype="auto",             # Use optimal data type for performance (e.g., float16)
    trust_remote_code=False         # Disable execution of remote custom code for security
)

# Load the corresponding tokenizer to handle text-token conversion
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline  # Import Hugging Face's pipeline utility for easy model inference

# Create a text generation pipeline using the loaded model and tokenizer
generator = pipeline(
    "text-generation",     # Task type: generate text (next-word prediction)
    model=model,           # Use the pre-loaded model (e.g., Phi-3)
    tokenizer=tokenizer,   # Use the corresponding tokenizer for text-token conversion
    return_full_text=False, # Return only the generated portion, not the input prompt
    max_new_tokens=500,    # Limit generation to 500 new tokens
    do_sample=False        # Use greedy decoding (no randomness); ensures deterministic output
)
```

```python
# The prompt (user input / query) formatted as a chat-style message
messages = [
    {"role": "user", "content": "Create a best joke on LLM's"}  # User's input prompt
]

# Generate output from the text-generation pipeline using the prompt
output = generator(messages)  # Pass the message(s) to the generator to produce a response

# Print the generated response text
print(output[0]["generated_text"])  # Access and print the generated text from the first (and only) result
```
